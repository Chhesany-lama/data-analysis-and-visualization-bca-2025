```python
import seaborn as sns

# Load the Iris dataset from seaborn (preloaded)
df = sns.load_dataset('iris')

# Documenting source
print("Dataset Source: Seaborn library – Iris dataset")
```

```python
import pandas as pd

# View structure
df.info()
df.head()

# Rename columns (if needed)
df.columns = [col.replace('_', ' ').title() for col in df.columns]

# Check for missing values
print(df.isnull().sum())
```

```python
# For simplicity, let’s filter only two species for clarity
filtered_df = df[df['Species'].isin(['setosa', 'versicolor'])]

print("Filtered to only two species for comparison: Setosa & Versicolor")
```

```python
# Basic statistics
print(filtered_df.describe())

# Distribution of species
print(filtered_df['Species'].value_counts())

# Correlation
correlation = filtered_df.corr(numeric_only=True)
print(correlation)
```

```python
import matplotlib.pyplot as plt

filtered_df.reset_index(inplace=True)

plt.figure(figsize=(10,5))
for species in filtered_df['Species'].unique():
    subset = filtered_df[filtered_df['Species'] == species]
    plt.plot(subset['index'], subset['Sepal Length'], label=species)

plt.title("Sepal Length Comparison")
plt.xlabel("Sample Index")
plt.ylabel("Sepal Length (cm)")
plt.legend()
plt.grid(True)
plt.show()
import matplotlib.pyplot as plt

filtered_df.reset_index(inplace=True)

plt.figure(figsize=(10,5))
for species in filtered_df['Species'].unique():
    subset = filtered_df[filtered_df['Species'] == species]
    plt.plot(subset['index'], subset['Sepal Length'], label=species)

plt.title("Sepal Length Comparison")
plt.xlabel("Sample Index")
plt.ylabel("Sepal Length (cm)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
for species in filtered_df['Species'].unique():
    subset = filtered_df[filtered_df['Species'] == species]
    plt.scatter(subset['Sepal Length'], subset['Petal Length'], label=species)

plt.title("Sepal vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.legend()
plt.show()
```

```python
import seaborn as sns

sns.pairplot(filtered_df, hue="Species", corner=True)
plt.suptitle("Pairwise Feature Comparison", y=1.02)
plt.show()
```

```python
# Example: Improved scatter with colorblind-safe palette and labels
palette = {'setosa': '#E69F00', 'versicolor': '#56B4E9'}

plt.figure(figsize=(8,6))
sns.scatterplot(data=filtered_df, x="Sepal Length", y="Petal Length", hue="Species", palette=palette)

plt.title("Refined Sepal vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend(title='Species')
plt.grid(True)
plt.tight_layout()
plt.show()
```
